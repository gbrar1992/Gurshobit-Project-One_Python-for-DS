### Ques 4. Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix.
    - Note: When the record/entry has "M"(indicates millions) as suffix you need to multiply that value with 1000000
    - When the record/entry has "K"(indicates thousands) as suffix you need to multiply that value with 1000

```python
## removing '€' symbol from column "Value"
fifaDataFrame.Value = [a.strip('€') for a in fifaDataFrame.Value]

## removing '€' symbol from column "Wage"
fifaDataFrame.Wage = [a.strip('€') for a in fifaDataFrame.Wage]

## removing '€' symbol from column "Release Clause"
fifaDataFrame['Release Clause'] = [a.strip('€') for a in (fifaDataFrame['Release Clause'].astype(str))]

fifaDataFrame
```

```python
## converting column "Value" to Float
## - multiply 1000000 for 'M' and 1000 for 'K'
fifaDataFrame.Value = [float(a.strip('K'))*1000 if 'K' in a else float(a.strip('M'))*1000000 for a in fifaDataFrame.Value]

fifaDataFrame.Value
```

```python
## converting column "Wage" to Float
## - multiply 1000000 for 'M' and 1000 for 'K'
fifaDataFrame.Wage = [float(a.strip('K'))*1000 if 'K' in a else float(a.strip('M'))*1000000 for a in fifaDataFrame.Wage]

fifaDataFrame.Wage
```

```python
## converting column "Release Clause" to Float
## - multiply 1000000 for 'M' and 1000 for 'K'
fifaDataFrame["Release Clause"] = [float(a.strip('K'))*1000 if 'K' in a else float(a.strip('M'))*1000000
                                    for a in (fifaDataFrame['Release Clause'].astype(str)) 
                                  ]
fifaDataFrame["Release Clause"]
```

```python
## dataset
fifaDataFrame
```

#### Summary of Code:
-     Currency symbol has been removed in columns namely: Value, Wage, Release Clause
-     All 3 columns are converted to Float datatype (dtype: float64)
-     'K' has been replaced by multiplying 1000
-     'M' has been replaced by multiplying 1000000 

###  Ques 5. Convert the column "Joined" into integer data type with keeping only the year

```python
## converting 'Joined' column data into date and getting year

from dateutil.parser import parse

fifaDataFrame.Joined = ['0' if 'nan' in a else parse(a).year for a in fifaDataFrame.Joined.astype(str)]
fifaDataFrame.Joined = fifaDataFrame.Joined.astype(int)

fifaDataFrame.Joined
```

#### Summary of Code:
-    Year has been extracted from the full date
-    "Joined" column is converted to Integer datatype (dtype: int32)

### Ques 7. The column 'Height' is in inches with a quotation mark, Convert to float with decimal points.

```python
## converting "Height" column to inches by removing quotation mark
fifaDataFrame.Height = ['0' if 'nan' in a else ((float(str(a).split("'")[0]) * 12) + (float(str(a).split("'")[1]))) 
                   for a in fifaDataFrame.Height.astype(str)]
fifaDataFrame.Height = fifaDataFrame.Height.astype(float)

fifaDataFrame.Height
```

#### Summary of Code:
-    "Height" column is converted to Float datatype (dtype: float64)
-    Quotation mark has been removed and now height is in inches

### Ques 8. The column "Weight" has the suffix as lbs, remove the suffix and convert to float.

### Ques 9. Check for the percentage of missing values and impute them with appropriate imputation techniques.

```python
## count total number of entries with missing values - column wise
fifaDataFrame.isnull().sum()
```

```python
## percentage of missing entries - column wise
for i in range(fifaDataFrame.shape[1]):
    col = fifaDataFrame.columns.values.tolist()[i]
    n_miss = fifaDataFrame[col].isna().sum()
    perc = n_miss / fifaDataFrame.shape[0] * 100
    print(' %d  %s , Missing: %d    (%.1f%%)' % (i, col, n_miss, perc))
```

```python
## including imputation for missing entries

## Imputation for column - float dataype
fifaDataFrame[["International Reputation","Weak Foot","Skill Moves","Jersey Number","Crossing","Finishing"]] = fifaDataFrame[["International Reputation","Weak Foot","Skill Moves","Jersey Number","Crossing","Finishing"]].fillna(0)
fifaDataFrame[["HeadingAccuracy","ShortPassing","Volleys","Dribbling","Curve","FKAccuracy","LongPassing"]] = fifaDataFrame[["HeadingAccuracy","ShortPassing","Volleys","Dribbling","Curve","FKAccuracy","LongPassing"]].fillna(0)
fifaDataFrame[["BallControl","Acceleration","SprintSpeed","Agility","Reactions","Balance","ShotPower"]] = fifaDataFrame[["BallControl","Acceleration","SprintSpeed","Agility","Reactions","Balance","ShotPower"]].fillna(0)
fifaDataFrame[["Jumping","Stamina","Strength","LongShots","Aggression","Interceptions","Positioning"]] = fifaDataFrame[["Jumping","Stamina","Strength","LongShots","Aggression","Interceptions","Positioning"]].fillna(0)
fifaDataFrame[["Vision","Penalties","Composure","Marking","StandingTackle","SlidingTackle","GKDiving"]] = fifaDataFrame[["Vision","Penalties","Composure","Marking","StandingTackle","SlidingTackle","GKDiving"]].fillna(0)
fifaDataFrame[["GKHandling","GKKicking","GKPositioning","GKReflexes","Release Clause"]] = fifaDataFrame[["GKHandling","GKKicking","GKPositioning","GKReflexes","Release Clause"]].fillna(0)

## Imputation for column - object datatype
fifaDataFrame[["Club","Preferred Foot","Work Rate","Body Type","Position","Loaned From"]] = fifaDataFrame[["Club","Preferred Foot","Work Rate","Body Type","Position","Loaned From"]].fillna('Missing')

## Imputation for column "Contract Valid Until" - datetime datatype
fifaDataFrame['Contract Valid Until'] = pds.to_datetime(fifaDataFrame['Contract Valid Until'], errors='coerce').ffill()


fifaDataFrame.info()
```

```python
## checking null value count for each column
fifaDataFrame.isnull().sum()
```

#### Summary of Code:
-    Total count of missing entries in each column is observed
-    Percentage of missing entries is calculated on column-basis
-    All null or NaN values are imputed as: 
        - 0 for integer and float datatype
        - 'Missing' for object datatype (string)
        - dummy date for datetime dataype
-    Finally, all column information is retrieved and it is verified that there are no null values in the dataset

### Ques 15. Generate a dataframe which should include the "Player name", "Club Name", "Wage", and 'Overall rating'.
    - i) find the average Overall rating for each club.
    - ii) Display the average overall rating of Top10 Clubs using a plot

```python
## generating dataframe
resultDataFrame = pds.DataFrame.copy(fifaDataFrame[['Name','Club','Wage','Overall']])
resultDataFrame
```

```python
## average Overall rating for each club
clubRatingAverage = resultDataFrame.groupby(['Club'])[['Overall']].mean()
clubRatingAverage.T
```

```python
## average overall rating of Top10 Clubs using a plot

best_dict = {}
for club in resultDataFrame['Club'].unique():
    overall_rating = resultDataFrame['Overall'][resultDataFrame['Club'] == club].mean()
    best_dict[club] = overall_rating
    
bestClub = pds.DataFrame.from_dict(best_dict, orient = 'index', columns = ['overall'])
bestClub['club'] = bestClub.index
bestClub = bestClub.sort_values(by = 'overall', ascending =  False)

matplt.figure(1 , figsize = (15 , 6))
sbn.barplot(x ='club', y ='overall', data = bestClub.head(10), palette = 'rocket')  
matplt.xticks(rotation = 30)

matplt.xlabel('Club', fontsize = 14)
matplt.ylabel('Average Overall Rating of players in club', fontsize = 14)
matplt.title('Clubs with best Players (average of overall ratings of players per club)', fontsize = 16)
matplt.ylim(0 , 100)

matplt.show()
```

#### Summary of Code & Output:
-    Dataframe is generated with data like "Player name", "Club Name", "Wage" and "Overall rating".

### Step 20. Generate a data frame containing top 5 players by Overall rating for each unique position. 

```python
## dataframe of top 5 players by Overall rating for each unique Position


def uniqueTopPosition(val):
    for pos in fifaDataFrame['Position'].unique():
        overallTopRating = fifaDataFrame['Overall'][fifaDataFrame['Position'] == pos].head(5)
        i = 0
        resultOneDataFrame = {}
        for i in range(5):
            resultOneDataFrame[i] = overallTopRating.index[i]
            i = i+1
        
        bestPositionDataFrameOne = pds.DataFrame.from_dict(resultOneDataFrame, orient='index', columns = ['index'])
        
        if str(val) is 'dataFrame':
            print('---------------------------------------------------------------------------------------------------------------')
            print('Dataframe of top 5 players based on Overall rating with Position : ' + pos)
            display(fifaDataFrame.loc[bestPositionDataFrameOne['index']])
            print('---------------------------------------------------------------------------------------------------------------')
            print()
            print()
        else:
            sampleDataFrameOne = fifaDataFrame.loc[bestPositionDataFrameOne['index']]
            print(sampleDataFrameOne.groupby('Position')[['Wage']].mean().T)
            print('----------------------')
    return


uniqueTopPosition('dataFrame')
```

#### Summary of Code & Output:
-    Dataframe is generated for the top 5 players on Overall Rating for each unique position: using a function.
-    There are a total of 28 unique positions (including 'Missing' position) in the players dataset.
-    Players are grouped based on positions and top 5 players are extracted and displayed for every position.
